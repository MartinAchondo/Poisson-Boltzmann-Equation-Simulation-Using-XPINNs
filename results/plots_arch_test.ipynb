```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x1 = [3,4,5,6,7,8,9,10,11]
fcnn_30 = [9.06e-3,3.189e-3,1.768e-3,1.62e-3,1.452e-3,3.386e-2,2.792e-3,1.557e-3,80]
fcnn_35 = [7.309e-3,2.5801e-3,1.304e-3,1.655e-3,2.4407e-3,3.2213e-1,2.48e-3,1.1278e2,1.024e-2]
x2 = [4,6,8,10,12,14,16,18]
resnet_30 = [3.1176e-3,1.24e-3,1.405e-3,1.27e-1,2.416e-1,2.984e-1,1.986e-2,8.625e1]
resnet_35 = [2.901e-3,1.17e-3,4.36e1,2.42e-2,9.355e-2,1.204e2,8.316e1,8.647e1]
```

```python
plt.plot(x1,fcnn_30, c='r', label='FCNN 30npl', marker='o', fillstyle='none')
plt.plot(x1,fcnn_35, c='g', label='FCNN 35npl', marker='o', fillstyle='none')
plt.plot(x2,resnet_30, c='b', label='ResNet 30npl', marker='o', fillstyle='none')
plt.plot(x2,resnet_35, c='m', label='ResNet 35npl', marker='o', fillstyle='none')
plt.legend()
plt.grid()
plt.yscale('log')
plt.xlabel('Hidden Layers')
plt.ylabel('Error')
plt.title("Total Loss");
```
