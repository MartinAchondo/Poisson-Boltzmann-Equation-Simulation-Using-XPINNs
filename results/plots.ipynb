```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 200
```

```python
dict_df = pd.read_excel('Results_Errors.xlsx', 
                   sheet_name=['20 per layer','30 per layer','40 per layer'])
```

```python
n20_df = dict_df .get('20 per layer')
n30_df = dict_df.get('30 per layer')
n40_df = dict_df.get('40 per layer')


print(n20_df)
print(n30_df)
```

Se corrieron 4000 iters

```python
x = n20_df['Hidden layers']
```

```python
plt.plot(x,n20_df['Total loss'], c='r', label='20 npl', marker='o', fillstyle='none')
plt.plot(x,n30_df['Total loss'], c='b', label='30 npl', marker='o', fillstyle='none')
plt.plot(x,n40_df['Total loss'], c='g', label='40 npl', marker='o', fillstyle='none')

plt.legend()
plt.grid()
plt.yscale('log')
plt.xlabel('Hidden Layers')
plt.ylabel('Error')
plt.title("Total Loss")
```

```python
plt.plot(x,n20_df['Interface loss'], c='r', label='20 npl', marker='o', fillstyle='none')
plt.plot(x,n30_df['Interface loss'], c='b', label='30 npl', marker='o', fillstyle='none')
plt.plot(x,n40_df['Interface loss'], c='g', label='40 npl', marker='o', fillstyle='none')

plt.legend()
plt.grid()
plt.yscale('log')
plt.xlabel('Hidden Layers')
plt.ylabel('Error')
plt.title("Interface Loss")
```
