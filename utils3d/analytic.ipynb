```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.grid()
```

```python
def border_value(x,y,z):
    sum = 0
    kappa = 0.125
    epsilon = 80
    pi=3.1415
    q = [(1,(0,0,0))]
    for qk,Xk in q:
        xk,yk,zk = Xk
        r_1 = np.sqrt((x-xk)**2+(y-yk)**2+(z-zk)**2)
        sum += qk/r_1*np.exp(-kappa*r_1)
    return (1/(epsilon*4*pi))*sum


def G_Fun(x,y,z):
    # epsilon es del interior
    q = [(1,(0,0,0))]
    pi=3.1415
    epsilon_G=2
    sum = 0
    for qk,Xk in q:
        xk,yk,zk = Xk
        r_1 = 1/((x-xk)**2+(y-yk)**2+(z-zk)**2)
        sum += qk*np.sqrt(r_1)
    return (1/(epsilon_G*4*pi))*sum
```
