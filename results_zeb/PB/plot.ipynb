```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
q = 2
e1 = 2
e2 = 80
k = 0.125
R = 1
```

```python
r1 = np.linspace(0.005,1,100)
r2 = np.linspace(1,10,100)

u1 = q/(4*np.pi)*(1/(e1*r1)-1/(e1*R)+1/(e2*R*(1+k*R)))

u2 = q*np.exp(-k*(r2-R))/(4*np.pi*e2*r2*(1+k*R))
```

```python
plt.plot(r1,u1)
plt.plot(r2,u2)
plt.grid()
```
